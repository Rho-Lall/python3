# Wave File Pre-Processing

This file takes a long .wav file and divides it into small 10 second pieces.

We did this to make the data set more uniform in size and to create multiple data points: instead of looking at one minute long file, we choose to look at 6 ten second files.

## Install the Library, PyDub

```python
!pip install pydub
from pydub import AudioSegment
from pydub.utils import make_chunks
import wave
print('done')
```

## Identify the file and filepath

```python
filename = 'g_sharp_mono'
filetype = '.wav'
filepath = 'C:/Users/rho/@ Code/'

file = filepath + filename + filetype
file
```

# Split up the files.

```python
myaudio = AudioSegment.from_file(file , "wav") 
```

```python
chunk_length_ms = 1000 # pydub calculates in millisec
chunks = make_chunks(myaudio, chunk_length_ms) #Make chunks of one sec

#Export all of the individual chunks as wav files

for i, chunk in enumerate(chunks):
    chunk_name = "sharp{0}.wav".format(i)
    print ("exporting", chunk_name)
    chunk.export(chunk_name, format="wav")
print("done")
```

## Fourier Transformation via MatPlotLib pColorMesh

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from scipy.io import wavfile

print('done')
```

```python
def scatterplotgenerator(file_arg):
  
  name = file_arg.split('.')
  savedfilename = "{0}.png".format(name[0])
  
  sample_rate, samples = wavfile.read(file_arg)
  frequencies, times, spectrogram = signal.spectrogram(samples, sample_rate)
  plt.pcolormesh(np.log(spectrogram))
  #plt.imshow(spectrogram)
  #plt.specgram(samples,Fs=sample_rate)
  #plt.ylabel('Frequency [Hz]')
  #plt.xlabel('Time [sec]')
  plt.savefig(savedfilename)
  #plt.show
  

#scatterplotgenerator('real0.wav')
```

```python
chunk_max = 54

for i in range(chunk_max+1):
  chunkfile = filepath + "sharp{0}.wav".format(i)
  scatterplotgenerator(chunkfile)
```
