# Prepare Image Data for Deep Learning

## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2

print("Done.")
```

# Trandform Images: Grey Scale & Re-size

```python
DATADIR = "C:/Users/rho/@ Code/images/trainingImages"
KEYS = ['g_normal', 'g_flat', 'g_sharp']

for key in KEYS:
    path = os.path.join(DATADIR, key) # path to the files
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array, cmap="gray")
        plt.show()
        break
    break
```

```python
img_array.shape
```

```python
LENGTH = 432
HEIGHT = 288

new_array = cv2.resize(img_array, (LENGTH, HEIGHT))
plt.imshow(new_array, cmap='gray')
plt.show()
```

## Create Tensor object

```python
training_data = []

def create_training_data():
    for key in KEYS:
        path = os.path.join(DATADIR, key) # path to the files
        class_num = KEYS.index(key)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
                new_array = cv2.resize(img_array, (LENGTH, HEIGHT))
                training_data.append([new_array, class_num])
            except Exception as e:
                pass

        
create_training_data()
```

```python
print(len(training_data))
```

```python
import random

random.shuffle(training_data)
```

```python
X = []
y= []
```

```python
for features, label in training_data:
    X.append(features)
    y.append(label)

X = np.array(X).reshape(-1, LENGTH, HEIGHT, 1)
```

```python
import pickle

pickle_out = open("X.pickle", "wb")
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open("y.pickle", "wb")
pickle.dump(y, pickle_out)
pickle_out.close()
```
